## Interpolação de Lagrange

```python
import sympy as sy
import numpy as np
from scipy.interpolate import lagrange
from numpy.polynomial.polynomial import Polynomial
from scipy.interpolate import lagrange
x=[1.3,1.4,1.5]
y=[3.669, 4.055,4.482]
x = np.array(x)
y = np.array(y)
poly = lagrange(x, y)
Polynomial(poly).coef
```

```python
poly(1.32)
```

```python
import matplotlib.pyplot as plt
p = np.poly1d(np.polyfit(x, y, 2))
t = np.linspace(0, 4, 20)
plt.plot(x, y, 'o', t, p(t), '-')
plt.show()
```

## Interpolação Linear com todos os pontos usando polyfit e poly1d

```python
import sympy as sy
import numpy as np
from scipy.interpolate import lagrange
from numpy.polynomial.polynomial import Polynomial
x=[0,1,2,3,4]
y=[32,47,65,92,132]
x = np.array(x)
y = np.array(y)
# valor 1 representa o grau do polinomio
z = np.polyfit(x, y, 1)
print(z)
p = np.poly1d(z)
p(3.7)
```

## Interpolação Linear com dois pontos

```python
x=[3,4]
y=[92,132]
x = np.array(x)
y = np.array(y)
z = np.polyfit(x, y, 1)
print(z)
p = np.poly1d(z)
p(3.7)
```

## Interpolação com interp

```python
import numpy as np
x=[0,1,2,3,4]
y=[32,47,65,92,132]
print(np.interp(3.5, x, y))
print(np.interp(5, x, y))
```

```python
import sympy as sy
import numpy as np
from scipy.interpolate import lagrange
from scipy.interpolate import lagrange
from numpy.polynomial.polynomial import Polynomial
x=[16, 40, 64]
y=[4.1, 0.6, 0.3]
x = np.array(x)
y = np.array(y)
poly = lagrange(x, y)
Polynomial(poly).coef
```

## Usando todos os pontos

```python
x=[16, 40, 64, 88, 112]
y=[4.1, 0.6, 0.3, 3.2, 8.6]
x = np.array(x)
y = np.array(y)
z = np.polyfit(x, y, 4)
print(z)
p = np.poly1d(z)
p(48)
```
