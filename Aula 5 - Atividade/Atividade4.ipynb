**Exercício 1:**

Resolver o seguinte sistemas linear:
$$
\left\{
\begin{align}
& x-2y+z+3t=0\\
& 2x+2y-z+t=0\\
& -x-y+3z+2t=0\\
& x-8y-z+3t=0
\end{align}
\right.
$$

Responder:\
SPD para sistema possível e determinado\
SPI para sistema possível e indeerminado\
SI para sistema impossível
          

```python
import numpy as np
x=[]

def gaussElimin(C,n,m):
    for k in range (0,n):
        if C[k,k]==0:
            for s in range (k+1,n):
                if C[s,k] != 0:
                    aux = C[k,:]
                    C[k,:] = C[s,:]
                    C[s,:]=aux
        for i in range (k+1,n):
            mik=C[i,k]/C[k,k]
            for j in range (k,m):
                C[i,j]=C[i,j]-mik*C[k,j]

    return C

A=np.array([[1,-2,1,3,0],[2,2,-1,1,0],[-1,-1,3,2,0],[1,-8,-1,3,0]])
#transfoma em float
A=A.astype(float)
n = len(A)
m = len(A[0])
X= gaussElimin(A,n,m)
if X[n-1,m-2]==0 and X[n-1,m-1]==0:
    print('SPI')
elif X[n-1,m-2]==0 and [n-1,m-1]!=0:
    print('SI')
else:
    x = n*[0]
    for i in range(n-1, -1, -1):
        s = sum([A[i][j]*x[j]for j in range(i+1,n)])
        x[i] = (A[i][n] - s)/A[i][i]
    print("SPD")


from sympy import Matrix, solve_linear_system
from sympy.abc import x, y,z, t
system = Matrix(((1,-2,1,3,0),(2,2,-1,1,0),(-1,-1,3,2,0),(1,-8,-1,3,0)))
solve_linear_system(system, x, y, z, t)
```
